```python
#Tugas Praktikum No 1 
#Modul 6 
#M Farhan Ramadhan ; 2010441005

import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 

C = [35,40,41,55,52,37,46,48] #------> Nilai x 
Q = [4050,6075,5400,9500,7290,5700,6210,8440] #-----> Nilai y 

n = len(C)


#Mencari Nilai komponen pada rumus 
sum_C = 0
sum_Q = 0
sum_QC= 0
sum_C2= 0

qc= []
c2= []
no= []
for i in range (1,n+1):
  no.append(i)

for i in range (n):

  sum_C += C[i]
  sum_Q += Q[i]


  QC = C[i]*Q[i]
  C2 = (C[i])**2

  sum_QC += QC
  sum_C2 += C2 
  qc.append(QC)
  c2.append(C2)

#Mencari nilai  a1 dan a0 
a1 = ((n*sum_QC)-(sum_Q*sum_C)) / ((n*sum_C2)-(sum_C**2))
print('Nilai a1 adalah {0:.3f}'.format(a1))

a0 = (sum_Q/n) - (a1*(sum_C/n))
print('Dan Nilai a0 yang kita dapatkan adalah {0:.2f}'.format(a0))

print('\nPersamaan garis yang kita dapatkan dari persamaan tersebut adalah ')
print('=====> y = {:.2f} + {:.3f}*x\n'.format(a0,a1))

#Tabel untuk nilai dari komponen 
data = {'C':C,
        'Q':Q,
        'Q*C':qc,
        'C^2':c2}
df = pd.DataFrame(data,index=no)
print(df)
print('========')
print('Jumlah :')
print('Jumlah C   = ',sum_C)
print('Jumlah Q   = ',sum_Q)
print('Jumlah Q*C = ',sum_QC)
print('Jumlah C^2 = ',sum_C2)
print()

#Memplot Grafik dari Data dan Persamaan yang didapatkan 
y=[]
for i in range (n):
  a = -3138.80 + 219.704*C[i]
  y.append(a)

plt.title('Grafik Data Regresi Linear') # =====> Judul dari tabel
plt.plot(C,Q,'o')                       # =====> Data sebelum regresi 
plt.plot(C,y,color='red')               # =====> Data Regresi Linear 
plt.legend(['Data Sebelum Regresi','Data Regresi Linear'])
plt.grid()
plt.show()
print('\n TABEL KOMPONEN DATA ')
print('=====================')

df.head()
```

```python
#Tugas Praktikum No 2 
#Modul 6 
#M Farhan Ramadhan ; 2010441005

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Persamaan awal nya adalah Miu = De^(B/T)
#Karena persamaan berupa pangkat, maka menggunakan ln pada persamaan

T = [0,5,10,20,30,40]
miu =[1.787,1.519,1.307,1.002,0.797,0.653]

n=len(T)

# ln y = ln miu 
# x = 1/T 
# ln a1 = D 
# Bet1 = B

sum_T = 0
sum_lnMiu = 0
sum_TMiu = 0
sum_T2 = 0

lnMiu = []
Tmiu  = []
t2    = []
no    = []

for i in range (n):
  sum_T +=T[i]

  sum_lnMiu += np.log(miu[i])
  lnMiu.append(np.log(miu[i]))

  TMiu = T[i]*(np.log(miu[i]))
  Tmiu.append(TMiu)

  T2 = T[i]**2
  t2.append(T2)
  
  sum_TMiu += TMiu
  sum_T2 += T2 

for i in range (1,n+1):
  no.append(i)

#Mencari nilai a1 dan b1 
B = ((n*sum_TMiu)-(sum_T*sum_lnMiu)) / ((n*sum_T2)-(sum_T**2))
print('Nilai b1 yang kita dapatkan adalah %0.4f' %B )

D = np.e**((sum_lnMiu/n) - B*(sum_T/n))
print('Nilai D yang kita dapatkan adalah {:.3f}'.format(D))

#Dan mendapatkan persamaan akhirnya sebagai berikut 
print('Maka persamaan akhir yang kita dapatkan adalah Miu = {:.3f}*e^({:.3f}/T) '.format(D,B))

#Mencari nilai pada T = 7,5 C
Tc = 7.5 
Miu = 1.719*(np.e**(-0.025/Tc))
print('Nilai miu pada Suhu 7,5 C adalah {:.3f}*10^-3 N.s/m2'.format(Miu))

#Mebuat tabel dari komponen rumus yang ingin dioperasikan 
data = {'T':T,
        'Miu':miu,
        'Ln Miu' :lnMiu,
        'T*LnMiu':Tmiu,
        'T^2':t2}

df = pd.DataFrame(data,index=no)
print()
print(df)
print()
#Memplot grafik dari data sebelum dan sesudah regresi linear
y = []
for i in range (n):
  yt = np.e**(np.log(D)+(B*T[i]))
  y.append(yt)

plt.title('Data Regresi Linear')
plt.plot(T,miu,'o')
plt.plot(T,y,color='red')
plt.legend(['Data Awal','Data Regresi Orde 2'])
plt.grid()
plt.show()
df.head()
```

```python
#Tugas Praktikum No 3 
#Modul 6 
#M Farhan Ramadhan ; 2010441005

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Pada no ini kita mencari persamaan regresi linear orde 2 
T = [0,5,10,20,30,40]
miu =[1.787,1.519,1.307,1.002,0.797,0.653]

n=len(T)

#Mencari komponen dari persamaan 

sum_T = 0
sum_T2= 0
sum_T3= 0
sum_T4= 0

sum_miu=0
sum_Tmiu=0
sum_T2miu=0


for i in range (n):
  sum_T += T[i]
  sum_T2+= (T[i]**2)
  sum_T3+= (T[i]**3)
  sum_T4+= (T[i]**4)

  sum_miu += miu[i]

  tmiu = T[i] * miu[i]
  sum_Tmiu += tmiu 

  t2miu = (T[i]**2) * (miu[i])
  sum_T2miu+=t2miu


#Membentuk matriks dari SPL 
print('A. Mendapatkan Persamaan Regresi Polinomial Orde 2 dari Data Tersebut')
print('=====================================================================\n')
print('o) Kita membentuk SPL menjadi bentuk Matriks ')
print('---------------------------------------------')
a = np.array([[n,sum_T,sum_T2,sum_miu],
             [sum_T,sum_T2,sum_T3,sum_Tmiu],
             [sum_T2,sum_T3,sum_T4,sum_T2miu]],float)
print(a)
N = len(a)
print()
print('o) Mendapatkan nilai a0, a1, a2 dari Eliminasi Gauss')
print('----------------------------------------------------')
for i in range (0,N-1):
  pivot = a[i][i]
  for j in range (i+1,N):
    c = a[j][i]/pivot
    for k in range (i,N+1):
      a[j][k] = a[j][k] - c*a[i][k]

print(a)
x =np.zeros((N,1))
x[N-1][0] = a[N-1][N]/a[N-1][N-1]

for i in  range (N-2,-1,-1):
  jum  = 0 
  for j in range (i+1,N):
    jum += a[i][j]*x[j][0]
    x[i][0] = (a[i][N]-jum)/a[i][i]
print()
print(x,'\n')    

#Kita dapatkan persamaan dalam Regresi Polinomial Orde 2 
print('o) Persamaan Akhir dari Regresi Polinomial Orde 2')
print('-------------------------------------------------')
print('y =   a0   +   a1*x   +   a2*x^2')
print('Persamaan yang kita adalah miu = {:.5f} + {:.5f}*T + {:.5f}*T^2\n'.format(float(x[0]),float(x[1]),float(x[2])))
print()


#Melalui persamaan diatas kita dapatkan nilai miu pada T = 7,5 C adalah 
print('B.Mendapatkan nilai miu pada T tertentu, (T = 7,5 C)')
print('====================================================')
t = 7.5
M = 1.76733 + -0.04952*(t) + 0.00055*(t**2)
print('''Nilai yang kita dapatkan dari persamaan regresi linear orde 2 
adalah {:.3f} N.s/m2'''.format(M))
```

```python
#Abaikan, dapat dari internet dan saya tidak ngerti juga 
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x, y):
    '''
    function to calculate the divided
    differences table
    '''
    n = len(y)
    coef = np.zeros([n, n])
    # the first column is y
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

def newton_poly(coef, x_data, x):
    '''
    evaluate the newton polynomial 
    at x
    '''
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p
#Mencari Persamaan Interpolas dengan Orde 4
I = np.array([0.25,0.75,1.25,1.5,2.0])
V = np.array([-0.45,-0.6,0.7,1.88,6])

#f(x) = V ; x = I
# get the divided difference coef
a_s = divided_diff(I, V)[0, :]

# evaluate on new data points
x_new = np.arange(0.25, 2.1, .1)
y_new = newton_poly(a_s, I, x_new)

plt.figure(figsize = (12, 8))
plt.plot(I, V, 'bo')
plt.plot(x_new, y_new)

```
