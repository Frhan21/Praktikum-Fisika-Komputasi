```python
#Modul 7 

import numpy as np

import matplotlib.pyplot as plt
x =5
y =5

colorinterpolation = 50 
colourMap = plt.cm.jet

#membuat matriks awal
T = np.zeros((x,y))
# syarat batas
T[(x-1):,:]=50 # bawah
T[:1,:]=300#atas
T[:,(y-1):]=100 #kanan
T[:,:1]=75# kiri

a = np.linspace(x,0,x)
b = np.linspace(y,0,y)
X, Y = np.meshgrid(b,a)

T_jac=T.copy()
print(T_jac)

maxt=1

iawal=0


#Persamaan ini cuman buat 1 kali iterasi 
T_jac[1:-1,1:-1]=0.25*(T[2:,1:-1]+T[:-2,1:-1]+T[1:-1,2:]+T[1:-1,:-2])
print(T_jac)

fig,ax=plt.subplots()
im = ax.imshow(T_jac)


for i in range(x):
  for j in range(y):
    text = ax.text(j,i,round(T_jac[i,j],1),ha='center',va='center',color='white')

ax.set_title("temperatur")
fig.tight_layout()

plt.subplots()
plt.contourf(X,Y,T_jac,colorinterpolation,cmap=colourMap)
plt.colorbar()



plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


B = 0.25
dt = 0.1 
dx = 2
t0 = 0 
tf = 10 

lamb = B*(dt/(dx**2))


colourinterpolation = 50 
colourmap = plt.cm.jet

x = 10 
y = 1 

n = round(x/dx)

T = np.zeros((n,n))
T[:,1:] = 50
T[1:,:1]= 0
T[:,-1:] =100
print(T)

for i in range (n-1):
  for j in range(1,n-1):
    T[i+1][j] = T[i][j] + lamb*(T[i][j+1]-2*T[i][j]+T[i][j-1])

print(T)

a = np.linspace(0,n,n)
b = np.linspace(0,n,n)

X,Y = np.meshgrid(a,b)
plt.contourf(X,Y,T,colourinterpolation,cmap='jet')
plt.colorbar()
plt.show()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

x = np.linspace(0,1,11)
n = len(x)
k = 5
dt = 1
dx = 0.1 
print(n)

U = np.sin((x*np.pi))
lamb = k * (dt**2)/(dx**2)

U[10] = 0
# print(U)

A = [0]
for i in range (1,n-1):
  a = lamb/2*(U[i+1]+U[i-1]) + (1-lamb)*U[i]
  A.append(a)

B = 0
A.append(B)


data = {'x':x,'U(x,0)':U,'U(x,1)':A}
df = pd.DataFrame(data)

print(df)

plt.plot(x,A,':<')
plt.xlabel('x')
plt.ylabel('U(x,1)')
plt.title ('Grafik Persamaan Gelombang Mekanik 1 Dimensi')
plt.legend(['d2U/dt2 = k2 d2U/dx2'])
plt.grid()
plt.show()
```
